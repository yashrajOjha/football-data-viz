```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch,VerticalPitch
```

```python
df = pd.read_csv('events.csv')
```

```python
df
```

```python
df['startX'] = df['startX']*1.2
df['startY'] = df['startY']*0.8
df['endX'] = df['endX']*1.2
df['endY'] = df['endY']*0.8
```

```python
fig,ax = plt.subplots(figsize=(13.5,8))
fig.set_facecolor('#000')
ax.patch.set_facecolor('#000')
pitch = Pitch(pitch_color='#000', line_color='#fff',positional=True, positional_color='#708090',positional_linewidth=0.5,
              stripe_color='#c2d59d', stripe=False)  # optional stripes
pitch.draw(ax=ax)

for i in range(len(df['startX'])):
    if df['outcome'][i] == 'su':
        plt.plot((df['startX'][i],df['endX'][i]),(df['startY'][i],df['endY'][i]),color='green',linewidth=2)
        plt.scatter(df['startX'][i],df['startY'][i],color='green')
    if df['outcome'][i] == 'un':
        plt.plot((df['startX'][i],df['endX'][i]),(df['startY'][i],df['endY'][i]),color='red',linewidth=3,alpha=0.75)
        plt.scatter(df['startX'][i],df['startY'][i],color='red')
        
plt.title("Doku vs Italy", color='white',size=20)
# plt.savefig('dokupassmap.png', bbox_inches='tight')
```
