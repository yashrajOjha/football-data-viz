```python
#import modules 
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch
```

```python
text_color ='w'
data = pd.read_csv('andresilva1.csv')
```

```python
data
```

```python
fig, ax = plt.subplots(figsize=(13,8.5))
fig.set_facecolor('#000')
ax.patch.set_facecolor('#000')

pitch = Pitch(pitch_type='statsbomb', orientation='vertical',
              pitch_color='#000000', line_color='#c7d5cc', figsize=(13,8),
              constrained_layout=False, tight_layout=True, view='half')
pitch.draw(ax=ax)
plt.gca().invert_yaxis()
for i in range(0,27):
    if data['shotType'][i] == 'LeftFoot' and data['xG'][i]<=0.5:
        plt.scatter(data['X'][i],data['Y'][i],s=50,c="#f5abc9")
    if data['shotType'][i] == 'LeftFoot' and data['xG'][i]>0.5:
        plt.scatter(data['X'][i],data['Y'][i],s=100,c="#f5abc9")
    if data['shotType'][i] == 'RightFoot'and data['xG'][i]<=0.5:
        plt.scatter(data['X'][i],data['Y'][i],s=50,c="#1eae98")
    if data['shotType'][i] == 'RightFoot'and data['xG'][i]>0.5:
        plt.scatter(data['X'][i],data['Y'][i],s=100,c="#1eae98")
    if data['shotType'][i] == 'Head'and data['xG'][i]<=0.5:
        plt.scatter(data['X'][i],data['Y'][i],s=50,c="#cf0000")
    if data['shotType'][i] == 'Head'and data['xG'][i]>0.5:
        plt.scatter(data['X'][i],data['Y'][i],s=100,c="#cf0000")
                    
            
# plt.title('Andre Silva goals in Bundesliga 20/21',fontsize='20', c='w')
plt.figtext(s=f'Andre Silva goals in Bundesliga 20/21',
        x=.52, y =.80, fontsize=16,fontfamily='Andale Mono',color='w')
plt.figtext(s=f'Goals: 27',
        x=.60, y =.65, fontsize=14,fontfamily='Andale Mono',color='w')
plt.figtext(s=f'xG: 23.2',
        x=.60, y =.60, fontsize=14,fontfamily='Andale Mono',color='w')
plt.figtext(s=f'LF-Goals:Pink',
        x=.60, y =.55, fontsize=14,fontfamily='Andale Mono',color='w')
plt.figtext(s=f'RF-Goals:Green',
        x=.60, y =.50, fontsize=14,fontfamily='Andale Mono',color='w')
plt.figtext(s=f'Headers:Red',
        x=.60, y =.45, fontsize=14,fontfamily='Andale Mono',color='w')
#plt.scatter(data['X'],data['Y'],s=100,c="#ea6969")
```
