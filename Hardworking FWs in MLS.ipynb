```python
from matplotlib import rcParams
rcParams['font.family'] = 'Century Gothic'
rcParams['font.sans-serif'] = ['Tahoma']
rcParams['font.weight'] = 'bold'
rcParams['font.size'] = '13'
rcParams['axes.labelsize']='medium'
rcParams['axes.titlesize']='large'
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("MLS.csv",encoding='ISO-8859-1') 
```

```python
len(df['Player'])
```

```python
df['Player'] = df['Player'].str.split('\\',expand=True)[0]
```

```python
fig, ax = plt.subplots(figsize=(13, 8))
fig.set_facecolor('#000')
ax.patch.set_facecolor('#000')
ax.xaxis.set_visible(True)
ax.yaxis.set_visible(True)
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')
ax.spines['bottom'].set_color('w')
ax.spines['left'].set_color('w')
plt.xlabel('Att 3rd Pressures p90',fontsize=10)
plt.ylabel('Mid 3rd Pressures p90',fontsize=10)
ax.xaxis.label.set_color('w')
ax.yaxis.label.set_color('w')
plt.axhline(y=np.median(df['Mid 3rd']), color="w", linestyle="dashed",alpha=0.5)
plt.axvline(x=np.median(df['Att 3rd']), color="w",linestyle="dashed",alpha=0.5)
ax.set_title('Hardworking forwards in the MLS', color='w',fontweight='bold',pad=30)
teams= ['New England','Nashville','Orland City','NYCFC','Philadelphia','CF Montral','D.C. United','Seattle','Colorado Rapids','Sporting KC','LA Galaxy','Minnesota Utd','Portland Timbers','Real Salt Lake']
for i in range(len(df['Player'])):
    if df['90s'][i] > 9.0:
        plt.scatter(df['Att 3rd'][i],df['Mid 3rd'][i],c='skyblue',s=90)
        if(df['Att 3rd'][i]> (np.median(df['Att 3rd'])) and df['Mid 3rd'][i]> (np.median(df['Mid 3rd']))):
            if(df['Squad'][i] in teams):
                ax.text(df['Att 3rd'][i]+0.01,df['Mid 3rd'][i]+0.01,df['Player'][i],c='w',fontsize=10)
plt.show()
```
